```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("../data/processed/churn_cleaned.csv")
```

```python
churn_counts = df["Churn"].value_counts()

plt.figure(figsize=(7, 7))
plt.pie(churn_counts,labels=["Yes", "No"],autopct="%1.1f%%",startangle=90,explode=(0.05, 0))

plt.title("Customer Churn Distribution (Yes / No)", fontsize=14)
plt.axis("equal")  
plt.show()
```

Through the above pie-chart, We can learn that most of the customers are churn in our dataset.

```python
df.groupby("ContractType")["MonthlyCharges"].mean().plot(kind="bar")
plt.ylabel("Mean price")
plt.xlabel("Contract type")
plt.title("Contract Type Average Price")
plt.show()
```

Contract Type vs Average Monthly Charges

The bar chart shows the average monthly charges across different contract types. Customers on month-to-month contracts have the highest average monthly charges, followed by one-year and two-year contracts. This suggests that long-term contracts are associated with relatively lower monthly costs, which may encourage customer retention. Customers on month-to-month plans may be more price-sensitive and therefore more likely to churn.

```python
plt.figure(figsize=(8, 5))
plt.hist(df["MonthlyCharges"],bins=30,edgecolor="black")
plt.title("Distribution of Monthly Charges", fontsize=14)
plt.xlabel("Monthly Charges")
plt.ylabel("Number of Customers")
plt.show()
```

Distribution of Monthly Charge

The histogram of monthly charges indicates a wide spread of customer billing amounts, ranging approximately from 30 to 120 units. The distribution is fairly uniform with slight concentration in the mid-to-high price range. Higher monthly charges may contribute to increased churn risk, as customers paying more may be more likely to discontinue services if they perceive insufficient value.

```python
plt.figure(figsize=(8, 5))
plt.hist(df["Tenure"],bins=30,edgecolor="black")
plt.title("Distribution of Customer Tenure", fontsize=14)
plt.xlabel("Tenure (Months)")
plt.ylabel("Number of Customers")
plt.show()
```

Distribution of Customer Tenure

The tenure distribution is right-skewed, showing that a large number of customers have low tenure (new or recent customers), while fewer customers remain with the service for a long period. This pattern indicates that churn is more common among early-stage customers, highlighting the importance of retention strategies during the initial months of subscription.
