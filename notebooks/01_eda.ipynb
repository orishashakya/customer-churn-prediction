```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("../data/processed/churn_cleaned.csv")
```

```python
df["Churn"].value_counts().plot(kind="pie")
plt.title("Churn(Yes/No)")
plt.ylabel("")
plt.show()
```

Through the above pie-chart, We can learn that most of the customers are churn in our dataset.

```python
df.groupby("ContractType")["MonthlyCharges"].mean().plot(kind="bar")
plt.ylabel("Mean price")
plt.xlabel("Contract type")
plt.title("Contract Type Average Price")
plt.show()
```

```python
plt.hist(df["MonthlyCharges"])
plt.title("Histogram of Monthly Charges")
plt.show()
```

##### Monthly Charges Distribution Summary

The histogram shows monthly charges are relatively uniformly distributed between $30-$120, with each price range containing 80-115 customers. There's a slight peak around $80-90, suggesting this may be a popular pricing tier. The continuous distribution indicates flexible pricing rather than rigid tiers, reflecting diverse customer segments from budget to premium. This uniform spread prompts further analysis: do lower-priced customers churn more due to less commitment, or do higher-priced customers churn more due to price sensitivity? A SQL query segmenting customers into pricing quartiles would help identify which price ranges have optimal retention rates.
